```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# # Example 1  Simple cost function.

$f(x) = xâˆš2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Make Data
x_1 = np.linspace (start = -3, stop = 3, num = 100)
x_1
```

```python
# Plot
plt.xlim ([-3, 3])
plt.ylim (0, 8)
plt.xlabel('x', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1))
plt.show()
```

## Slope and Derivatives 

```python
def df(x):
    return 2*x + 1
```

```python
# Plot function and derivative side by side.
plt.figure(figsize=[15, 5])

#1 Chart: Cost Function
plt.subplot(1, 2, 1)

plt.xlim ([-3, 3])
plt.ylim (0, 8)
plt.xlabel('x', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1))

#2 Chart Derivative
plt.subplot(1, 2, 2)
plt.plot(x_1, df(x_1))


plt.show()
```
